```python
import wordgraph as wg
import pandas as pd
from ggplot import *
from scipy import stats

bl_file = 'data/all_bl_no_fixme_no_CHI.csv'
G = wg.WordGraph()
```

```python
def merge_freq_edges(bl, g):
    df = pd.read_csv('data/all_bl_no_fixme_no_CHI.csv')
    bl = df['basic_level'].value_counts().to_frame()
    bl['word'] = bl.index
    bl.columns = ['freq', 'word']
    deg_dist = g.degree_distribution()
    df = pd.merge(bl, deg_dist, on='word')
    return df
```

```python
G.load_pickle('/Volumes/Lexar/semspace_output/seedlings_all/cos_0.13')
```

```python
df = merge_freq_edges(bl_file, G)
df.to_csv('output/seedlings_all_013_freq_edges.csv', index=False)
```

```python
rho, p = stats.spearmanr(df['degree'], df['freq'])
print("rho: {}\np:   {}".format(rho, p))
ggplot(df, aes(x='degree', y='freq')) +\
    geom_point()
```
