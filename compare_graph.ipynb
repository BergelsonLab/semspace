```python
seedlings_rho = pd.read_csv('output/edges_vs_cdi_seedlings_all.csv').query('(epsilon >= 0.63) & (epsilon <= 0.74)')
wordbank_rho = pd.read_csv('output/edges_vs_cdi_wordbank.csv').query('(epsilon >= 0.6) & (epsilon <= 0.7)')

seedlings_rho['dataset'] = 'seedlings'
wordbank_rho['dataset'] = 'wordbank'

data = seedlings_rho.append(wordbank_rho)
data.to_csv('output/seed_wordbank_overlapping_powerlaw.csv')
```

```python
p = ggplot(data, aes(x='epsilon', y='rho', color='dataset')) +\
    geom_point(size=40) +\
    ggtitle("Correlation at Overlapping Power Law Fits")
```
