```python
import networkx as nx
import json
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
with open("data/density6", "rU") as input:
    month6 = json.load(input)
```

```python
G = nx.MultiGraph()


# G.add_node("hello")
# G.add_node("my")
# G.add_node("name")
# G.add_node("is")
# G.add_node("andrei")

# G.add_edge("hello", "andrei")
# G.add_edge("andrei", "is")

for key in month6:
    G.add_node(key, size=1)
    
for key, value in month6.items():
    if value:
        for element in value: 
            G.add_edge(key, element[0])
```

```python
nx.draw(G, with_labels=False, node_size=4)

```

```python
f = plt.figure(figsize=(70, 50), dpi=1000)
nx.draw_spring(G, node_size=7, ax=f.add_subplot(111), with_labels=True, font_size=2, font_color="#009900", width=0.2, linewidths=0.2)
#nx.draw_graphviz(G, ax=f.add_subplot(111))

f.savefig("graph_euc.pdf")
```
