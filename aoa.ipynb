```python
import pandas as pd
from scipy.stats import *
import matplotlib.pyplot as plt
```

```python
wordbank = pd.read_csv("data/wordbank_english.csv")
```

```python
results = []
for row in wordbank.loc[:,'16':'30'].iterrows():
    found = False
    for index, value in enumerate(row[1].iteritems()):
        if value[1] >= 0.5:
            results.append(int(value[0]))
            found = True
            break
    if not found:
        results.append(30)
        
wordbank['reach_thresh'] = results
```

```python
pearsonr(wordbank['reach_thresh'], wordbank['16'])
```

```python
pearsonr(wordbank['reach_thresh'], wordbank['23'])
```

```python
pearsonr(wordbank['reach_thresh'], wordbank['30'])
```
